```python
# A sample of fineweb data from hugging face
with open(f"../data/fineweb_sample.txt", "r") as f:
    fineweb_sample = f.read()
```

```python
# A sequence of bits
# Raw binary bits of the sample - two symbols very large sequence
raw_bits = ''.join(format(b, '08b') for b in fineweb_sample.encode())
raw_bits
```

```python
# A sequence of bytes (group of 8)
# Convert them to list of numbers corresponding to each byte (8 bits from above) ranging from 0 -> 255
# shorter sequence (compared to prev) and more symbols 
bytes_seq = list(fineweb_sample.encode())
bytes_seq

```
